**Develop a text preprocessing and analysis application using NLTK for tokenization, POS
tagging, and basic NLP tasks**

```python
import nltk
import pandas as pd

# Essential downloads to prevent LookupErrors
resources = [
    'punkt',
    'punkt_tab',
    'stopwords',
    'averaged_perceptron_tagger_eng',
    'wordnet',
    'omw-1.4'
]

for res in resources:
    nltk.download(res)

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk.stem import WordNetLemmatizer
from collections import Counter

print("\n[SUCCESS] All NLTK resources are ready.")
```

```python
class TextProcessor:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()

    def analyze(self, text):
        # 1. Basic cleaning & Tokenization
        tokens = word_tokenize(text.lower())

        # 2. Filter out punctuation and stopwords
        clean_tokens = [w for w in tokens if w.isalnum() and w not in self.stop_words]

        # 3. Part-of-Speech Tagging
        # Tags words as Nouns (NN), Verbs (VB), etc.
        tags = pos_tag(clean_tokens)

        # 4. Lemmatization (Root word extraction)
        lemmas = [self.lemmatizer.lemmatize(w) for w in clean_tokens]

        # 5. Frequency Count
        counts = Counter(lemmas).most_common(5)

        return {
            "original": text,
            "tokens": clean_tokens,
            "tags": tags,
            "lemmas": lemmas,
            "top_freq": counts
        }
```

```python
# Initialize the app
nlp_app = TextProcessor()

# Input your text here
raw_data = """
Natural Language Processing allows computers to understand human speech.
NLTK is a leading platform for building Python programs to work with human language data.
It provides easy-to-use interfaces to over 50 corpora and lexical resources.
"""

# Run analysis
output = nlp_app.analyze(raw_data)

# Display a clean summary using a Pandas DataFrame for the Tags
print("--- TOP 5 FREQUENT LEMMAS ---")
print(output['top_freq'])

print("\n--- PART OF SPEECH TAGGING (Preview) ---")
df_tags = pd.DataFrame(output['tags'], columns=['Word', 'POS Tag'])
print(df_tags.head(10))
```
